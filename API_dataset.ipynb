```python
import json
import csv

# Load JSON file
with open('departures.json', 'r', encoding='utf-8') as file:
    data = json.load(file)

# Define CSV file name
csv_filename = 'departures.csv'

# Open CSV file and write data
with open(csv_filename, mode='w', newline='', encoding='utf-8') as file:
    writer = csv.writer(file)

    # Write header (assuming JSON is a list of dictionaries)
    if isinstance(data, list) and len(data) > 0:
        writer.writerow(data[0].keys())  # Write column names

        # Write rows
        for row in data:
            writer.writerow(row.values())

print(f"Data saved to {csv_filename}")
```
