```python
import pandas as pd

# Load CSV file
df = pd.read_csv("departures.csv")

# Print the first few rows
print(df.head())
```

```python

# Remove specific columns
df = df.drop(columns=["Year", "State"])

# Remove duplicates
df = df.drop_duplicates()

# Remove rows with missing values
df = df.dropna()

# Print cleaned data
print(df)
```

```python
# Remove the index column when saving to CSV
df.to_csv("cleaned_departures.csv", index=False)

# Load CSV file
dfclean = pd.read_csv("cleaned_departures.csv")

# Print the first few rows
print(dfclean.head())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd


# Sort by domestic flights and select top 20 for better visibility
dfclean = dfclean.sort_values(by="Domestic", ascending=False).head(20)

# Set figure size
plt.figure(figsize=(12, 8))

# Create the barplot
sns.barplot(x="Domestic", y="Name", data=dfclean, palette="Blues_r")

# Improve readability
plt.xlabel("Number of Domestic Flights")
plt.ylabel("Country Name")
plt.title("Top 20 Countries by Domestic Flights")
plt.xticks(rotation=45)  # Rotate x-axis labels if needed
plt.grid(axis="x", linestyle="--", alpha=0.7)

# Show the plot
plt.show()


plt.figure(figsize=(10, 6))
sns.barplot(x="International", y="Name", data=dfclean, palette="Reds_r")
plt.xlabel("Number of International Flights")
plt.ylabel("Country Name")
plt.title("International Flights by Country")
plt.grid(axis="x", linestyle="--", alpha=0.7)

plt.figure(figsize=(10, 6))
sns.barplot(x="Departures", y="Name", data=dfclean, palette="Greens_r")
plt.xlabel("Number of Departures Flights")
plt.ylabel("Country Name")
plt.title("Flights by Country")
plt.grid(axis="x", linestyle="--", alpha=0.7)



# Create boxplot
sns.boxplot(x=dfclean["Departures"], color="skyblue")
    
    # Labels and title
plt.xlabel("Number of Departures")
plt.title("Boxplot of Departures")
    
    # Show plot
plt.show()
```

```python
import numpy as np

plt.figure(figsize=(8, 5))
sns.boxplot(x=np.log1p(dfclean["Departures"]), color="skyblue")  # log1p handles zero values
plt.xlabel("Log(Number of Departures)")
plt.title("Boxplot of Departures (Log Scale)")
plt.show()

plt.figure(figsize=(8, 5))
sns.boxplot(x=dfclean["Departures"], color="skyblue")
plt.xlim(0, 1.5e6)  # Adjust range based on data distribution
plt.xlabel("Number of Departures")
plt.title("Boxplot of Departures (Adjusted Scale)")
plt.show()

Q1 = dfclean["Departures"].quantile(0.25)
Q3 = dfclean["Departures"].quantile(0.75)
IQR = Q3 - Q1

filtered_df = dfclean[(dfclean["International"] >= Q1 - 1.5 * IQR) & (dfclean["International"] <= Q3 + 1.5 * IQR)]

plt.figure(figsize=(8, 5))
sns.boxplot(x=filtered_df["Departures"], color="skyblue")
plt.xlabel("Number of International Flights")
plt.title("Boxplot of International Flights (Without Extreme Outliers)")
plt.show()

filtered_df = dfclean[(dfclean["Domestic"] >= Q1 - 1.5 * IQR) & (dfclean["Domestic"] <= Q3 + 1.5 * IQR)]

plt.figure(figsize=(8, 5))
sns.boxplot(x=filtered_df["Departures"], color="skyblue")
plt.xlabel("Number of Domestic Flights")
plt.title("Boxplot of Domestic Flights (Without Extreme Outliers)")
plt.show()

```
